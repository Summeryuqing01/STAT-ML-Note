* Non-linear Relationships: Unlike linear regression which fits a straight line, it fits a polynomial equation to capture the curve in the data. A linear model often results in residuals with noticeable patterns which shows a poor fit.
* Flexibility and Complexity: It does not assume all independent variables are independent. By introducing higher-degree terms, it allows for more flexibility and can model more complex, curvilinear relationships between variables.

Advantages: 
1. Fits a Wide Range of Curves: Polynomial regression can model a broad range of non-linear relationships helps in making it versatile for complex data.
2. Captures Non-linear Patterns: It provides a more accurate approximation of relationships when data follows a curvilinear pattern, unlike linear regression.
3. Flexible Modeling: It can capture and represent the nuances in the data helps in making it ideal for situations where linear models fail.

Disadvantages:
1. Sensitivity to Outliers: It is highly sensitive to outliers and a few extreme values can skew the results significantly.
2. Overfitting: With higher-degree polynomials, thereâ€™s a risk of overfitting where the model becomes too complex and fails to generalize well to new data.
3. Limited Outlier Detection: Unlike linear regression, it has fewer built-in methods for detecting or handling outliers helps in making it challenging to identify when the model is affected by them.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Dataset
X = 8 * np.random.rand(500, 1)
y = 1 + X**2+X + 2 + np.random.randn(500,1)
# Plot
plt.scatter(X,y, alpha=0.5);
```

```python
# Data distribution
plt.title('Salary Distribution Plot')
sns.histplot(y)
plt.show()
```

```python
# Train linear regression model on whole dataset
X_train = np.array(X[0:-20]).reshape(-1, 1)
y_train = np.array(y[0:-20]).reshape(-1, 1)
lr = LinearRegression()
lr.fit(X_train, y_train)

# Train polynomial regression model on the whole dataset
pr = PolynomialFeatures(degree = 2, include_bias = False)
X_poly = pr.fit_transform(X_train)
lr_2 = LinearRegression().fit(X_poly, y_train)
```

```python
# Predict results
y_pred_lr = lr.predict(X_train)           # Linear Regression
y_pred_poly = lr_2.predict(X_poly)  # Polynomial Regression
```

```python
# Visualize real data with linear regression
plt.scatter(X, y, color = 'blue',alpha = 0.5)
plt.plot(X_train, y_pred_lr, color = 'firebrick')
plt.title('Real data (Linear Regression)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['X/y_pred_lr', 'X/y'], loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
# y_mean
label_mean =  np.mean(y)
print('label mean:', label_mean )
# RMSE
rmse = np.sqrt(mean_squared_error(y_train,y_pred_lr))
print('RMSE:', rmse )
# % Off
print('% off:',rmse/label_mean)
```

```python
# Visualize real data with polynomial regression
plt.scatter(X,y, color = 'blue',alpha=0.5)
plt.plot(X_train, y_pred_poly, color = 'firebrick',linestyle='', marker='.', lw=0.1)
plt.title('Real data (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['X/y_pred_poly', 'X/y'], loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# RMSE
rmse = np.sqrt(mean_squared_error(y_train,y_pred_poly))
print('RMSE:', rmse )
# % Off
print('% off:',rmse/label_mean)
```
